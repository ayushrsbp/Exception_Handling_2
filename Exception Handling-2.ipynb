```python
# Q3.
# Errors defined in the ArithmeticError class are following:
#     1. FloatingPointError
#     2. OverflowError
#     3. ZeroDivisionError
#     4. DecimalException

#     * OverflowError:
#         It typically occurs when performing computations that generate numbers that are too large to be stored or represented accurately.
        
def get_value(a, b):
    try:
        result = a ** b
    except OverflowError as e:
        print("Error:", e)

# In the above example, when the value of a ** b will exceed the maximum representable value for a numeric type, the exception is caught in 'except OverflowError' block, and appropriate error is printed.

    # * ZeroDivisionError:
    #     It occurs when a number is divided by zero.
    
def get_value(a, b):
    try:
        result = a/b
    except OverflowError as e:
        print("Error:", e)
        
# In the above example, when the value of b is equal to 0, the exception is caught in 'except ZeroDivisionError', and an appropriate error is printed.
```
